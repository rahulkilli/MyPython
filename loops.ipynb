```python
original='Abcd@1234'
print("Welcome!")
print("-*"*20)
user_pass=input("Please enter password ")
count=1
while user_pass!=original and count<=2:
  user_pass=input("Please enter password ")
  print("Wrong Password!")
  print("Try again")
  count+=1
  if user_pass==original:
    print("Hurray! Welcome")
    break
  if count==3:
    print("Forgot password")
    break
print("Welcome")
```
