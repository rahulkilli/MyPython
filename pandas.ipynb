# Series

> One dimensional data; like column of a table

```python
l=[1,2,3,4,5]
s=pd.Series(l)
print(s)
```

```python
s=pd.Series(["Apple","Banana",'Mango'])
print(s)
```

```python
t=pd.Series(["Name",20,"Male","Tech"])
print(t)
```

```python
s=pd.Series(["Apple","Banana",'Mango'],index=["A","B","C"])
print(s)
```

```python
gender=pd.Series(["Male","Female","Male","Female","Female"],index=["a","b","c","d","e"])
print(gender)
```

```python
gender=pd.Series(["Male","Female","Male","Female","Female"],index=["a","b","c","d","f"])
print(gender)
print()
print(type(gender))
```

```python
d={
    "A":90,
    "B":93,
    "C":75,
    "D":78,
    "E":95
}
marks=pd.Series(d)
print(marks)
```

```python
d={
    "A":90,
    "B":93,
    "C":75,
    "E":78,
    "F":95
}
marks=pd.Series(d,index=["A",2,"C","D",5])
print(marks)
```
